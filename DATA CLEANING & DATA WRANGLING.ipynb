## Placing 'part_id2' at the beginning of the dataset

I stored the column in variable col_partid2, then dropped 'part_id2' from the original dataset, which is shown below

The new dataset is stored in new_df

```python
col_partid2 = df['part_id2']

new_df = df.drop(columns = 'part_id2')
new_df
```

By using the concat() we can now add the column 'part_id2' back into the beginning of the dataset

```python
new_df = pd.concat([col_partid2, new_df], axis = 1)
new_df
```

## Sorting the data by the 'part_id2' column (smallest to largest)

By using sort_values() we can specify to sort by column 'part_id2' and use the ascending attribute

```python
new_df = new_df.sort_values(by = 'part_id2', ascending = True)
new_df
```

## Identifying all the missing data 

several ways to find missing information (total, per column, per row)

isna() returns missing values in boolean (True = missing)

## Total
By using sum() twice we add the missing values of the entire dataset

This can be put in a sentence to make results easier to understand like shown below

```python
total = new_df.isna().sum().sum()
missing_data = "Total number of cells with missing data: {}".format(total)
missing_data
```

## Per column



```python
per_col = pd.DataFrame(new_df.isna().sum())
per_col.rename(columns = {0:'Missing Data Per Row'})
```

## Per row

```python
per_row = pd.DataFrame(new_df.isna().sum(axis = 1))
per_row.rename(columns = {0:'Missing Data Per Row'})
```

## Recoding the gender variable so that it is no longer a string variable
can be changed to an integer

F changed to 1

M changed to 0

```python
new_df['gender'] = new_df['gender'].replace({'F': 1, 'M': 0})
new_df
```

## Dropping other columns like random and insample

```python
new_df = new_df.drop(columns = ['random', 'insample'])
new_df
```

## Finding the mean, standard deviation, and range of each column
By storing each calculation as a DataFrame, it will make it easier to combine it into one table and display the results

```python
mean = pd.DataFrame(new_df.mean())
st_dev = pd.DataFrame(new_df.std()) 
range_ = pd.DataFrame(new_df.max() - new_df.min())
```

Rename each DataFrame

```python
mean = mean.rename(columns = {0:'Mean'})
st_dev = st_dev.rename(columns = {0:'Standard Deviation'})
range_ = range_.rename(columns = {0:'Range'})
```

We can now combine them into one table

The table with the results is neat, organized and readable.

```python
frames = [mean, st_dev, range_ ]
table = pd.concat(frames, axis = 1) 
table
```

## Finding the mean, standard deviation and range using gender as the grouping variable. 

First isolate the data 

Then find all necessary calculations

Put results in a easy to read table

```python
female = new_df[new_df['gender'] == 1]

fem_mean = female.mean()
fem_std = female.std()
fem_range = female.max() - female.min()

fem_frame = pd.DataFrame([fem_mean,fem_std, fem_range], index=['Mean', 'Standard Deviation', 'Range'])
title_fem = fem_frame.index.name = 'Female'
fem_frame
```

```python
male = new_df[new_df['gender'] == 0]

m_mean = male.mean()
m_std = male.std()
m_range = male.max() - male.min()

m_frame = pd.DataFrame([m_mean,m_std, m_range], index=['Mean', 'Standard Deviation', 'Range'])
title_male = m_frame.index.name = 'Male'
m_frame
```

## Replacing missing data with the mean, min, max and zero

```python
average = new_df.fillna(new_df.mean())
minimum = new_df.fillna(new_df.min())
maximum = new_df.fillna(new_df.max())
zero = new_df.fillna(0)
```
